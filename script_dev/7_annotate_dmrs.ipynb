```python
# Determine if there is overlap between genes and DMRs
# query and reference
def overlap(q_st, q_end, ref_st, ref_end):
    overlap  = min(q_end, ref_end)-max(q_st, ref_st)
    return overlap
```

```python
def gencode_annotate_loci(a, tb):
    genes = []

    try:
        for region in tb.fetch(a['chr'], int(a['start']), int(a['end'])):
            if region:
                r = region.split('\t')
                overlap_len = overlap(int(a['start']), int(a['end']), int(r[1]), int(r[2]))
                
                ### Percentage of the input region that overlap with the feature
                ret_val = '{}({})'.format(r[4], np.round(overlap_len/float(int(a['end'])-int(a['start']))*100, 2)) 
                
                ### Percentage of the feature that overlaps with the region
                # If you want to find the fraction of the GENCODE gene that overlaps with the input interval 
                # (e.g. overlap_len/float(int(r[2])-int(r[1]))*100).
                # overlap_percent = '{}({})'.format(r[4], np.round(overlap_len/float(int(r[2])-int(r[1]))*100, 2)

                genes.append(ret_val)

                feature = r[3] 
                gene_type = r[5]
                gene_level = r[6]
                exon_id = r[7]
                exon_number = r[8]
                transcript_name = r[9]
                transcript_type = r[10]
                 

        if len(genes)>0:
            return [";".join(genes), feature, gene_type, gene_level, exon_id, exon_number, transcript_name, transcript_type]
        else:
            return ["NA(0)", np.nan, np.nan, np.nan, np.nan, np.nan, np.nan, np.nan]
    except ValueError:
        return ["NA(0)", np.nan, np.nan, np.nan, np.nan, np.nan, np.nan, np.nan]
    
# # loci is a row in a dataframe with columns chr, start, end ['#chr', 'start', 'end']
# # tabix is the indexed annotation file as a pysam.TabixFile object
# def gencode_annotate_loci(loci, tabix):
#     genes = []
#     try:
#         # reads the tabix file
#         for region in tabix.fetch(loci['chr'], int(loci['start']), int(loci['end'])):
#             if region:
#                 r = region.split('\t')

#                 ### Length of the overlap between the query interval and the gencode gene
#                 # 
#                 overlap_len = overlap(int(loci['start']), int(loci['end']), int(r[1]), int(r[2]))

#                 ### Percentage of the query interval that overlap with the gencode gene
#                 # If you want to find the fraction of the GENCODE gene that overlaps with the input interval 
#                 # (e.g. overlap_len/float(int(r[2])-int(r[1]))*100).
#                 query_overlap_percent = '{}({})'.format(r[3], np.round(overlap_len/float(int(loci['end'])-int(loci['start']))*100, 2)) 
#                 genes.append(query_overlap_percent)
#         if len(genes)>0:
#             return ";".join(genes)
#         else:
#             return "NA(0)"
#     except ValueError:
#         return "NA(0)"
```

```python
# annotate dmrs using gencode_annotate_loci function and gencode_v44 annotation file
# returns genes, feature, gene_type, gene_level, exon_id, exon_number, transcript_name, transcript_type that overlap with DMRs
# returns amount of overlap as a percentage under gene column as gene-name(overlap %)
df[['gene', 'feature', 'gene_type', 'gene_level', 'exon_id', 'exon_number', 'transcript_name', 'transcript_type']] = df.apply(lambda x: gencode_annotate_loci(x[['chr', 'start', 'end']], gencode_v44), axis=1, result_type='expand')

# replace empty strings with NaN
df = df.replace(r'^\s*$', np.nan, regex=True)
```
